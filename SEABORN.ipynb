```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.barplot(x="sex",y="total_bill",data=tips,color="black",estimator=np.std) #by using estimator graph show w.r.t std deviation 
plt.show()                                                              
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.boxplot(x="day",y="tip",data=tips,hue="smoker")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
sns.kdeplot(tips["total_bill"])
```

```python
sns.displot(tips['total_bill'],kde=False,bins=10)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex')
```
