### Characteristics

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y1=x**2
plt.plot(x,y,color="black",marker="+",linewidth=0.5,label="x^3")
plt.plot(x,y1,color="red",marker="*",linewidth=1,label="x^2")
plt.legend()
plt.grid()
plt.axis([0,10,0,150])
plt.xlabel("x")
plt.ylabel("x^3")
plt.title("Matplotlib Demo")
plt.text(5,100,"Graph",fontsize=12)
plt.show()
```

### Bubble Chart

```python
import matplotlib.pyplot as plt
year=[2012,2014,2016,2018,2020]
score=[12,15,18,21,24]
population=[100,120,150,200,300]
plt.scatter(year,score,s=population,c='r',alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
year=[2012,2014,2016,2018,2020]
score=[12,15,18,21,24]
population=[100,120,150,200,300]
import numpy as np
c=np.arange(len(year))
c[0]=c[4]
plt.scatter(year,score,s=population,c=c,alpha=1,edgecolor='b',marker="^")
plt.show()
```

### Pie Chart

```python
labels=['A','B','C','D']
sizes=[2,4,8,16]
colors=['red','yellow','pink','orange']
explode=[0.1,0,0,0]
plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct="%.1f%%",counterclock=False,startangle=180)
plt.axis("equal")
plt.show()
```

### Histogram

```python
a=[1,2,1,3,3,4,2,5,6,7,10,6,20,22,11,14]
xt=np.arange(23)
plt.hist(a,edgecolor='black',bins=21)
plt.xticks(xt)
plt.show()
```
